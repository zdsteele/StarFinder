```python
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os
```

```python
# Where to save the figures
PROJECT_ROOT_DIR = "."
# CHAPTER_ID = "end_to_end_project"
IMAGES_PATH = os.path.join(PROJECT_ROOT_DIR, "static")
os.makedirs(IMAGES_PATH, exist_ok=True)

def save_fig(fig_id, tight_layout=True, fig_extension="png", resolution=300):
    path = os.path.join(IMAGES_PATH, fig_id + "." + fig_extension)
    print("Saving figure", fig_id)
    if tight_layout:
        plt.tight_layout()
    plt.savefig(path, format=fig_extension, dpi=resolution)
```

```python
from sklearn.model_selection import train_test_split

train_set, test_set = train_test_split(df, test_size=0.2, random_state=42)
```

```python
test_set['class'].value_counts()
```

```python
# Review Stratified Data
df["redshift"].hist()
```

```python
df['redshift_cat'] = pd.cut(df['redshift'],
                            bins = [-6, 1, 2, 3, 4, 5, 10],
                            labels=[-1,.5,1,2,3,4])
```

```python
df['redshift_cat'].hist()
```

```python
from sklearn.model_selection import StratifiedShuffleSplit

split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
for train_index, test_index in split.split(df, df['redshift_cat']):
    strat_train_set = df.loc[train_index]
    strat_test_set = df.loc[test_index]
```

```python
df
```

```python
for set_ in (strat_train_set, strat_test_set):
    set_.drop("redshift_cat", axis=1, inplace=True)
```
